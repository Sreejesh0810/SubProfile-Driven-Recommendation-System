#MF ALS

```python
!pip install implicit
from implicit.als import AlternatingLeastSquares
from scipy.sparse import csr_matrix
```

This Class is use to train Mf (ALS) Model and genrate Recommend Set For Computing Diversity Score

```python
class Matrix_Factorization_ALS:

  def calc_avg(self,data):
    self.avg_rating = data.ratings_data['rating'].mean()
    return self.avg_rating

  # Method to calculate item bias
  def calc_item_bias(self,data):
    item_counts = data.ratings_data["movieId"].value_counts()
    item_ratings =data.ratings_data.groupby("movieId")['rating'].sum()
    self.bi = (item_ratings / item_counts - self.avg_rating).fillna(0)
    return self.bi
  # Method to Calculate user bias
  def calc_user_bias(self,data):
    # Compute user biases
    user_counts = data.ratings_data["userId"].value_counts()
    user_ratings = data.ratings_data.groupby("userId")['rating'].sum()
    self.bu = (user_ratings / user_counts - self.avg_rating).fillna(0)
    return self.bu

  def calc_als(self,user_item,factors=50, regularization=0.01, iterations=100):
    # Initialize ALS model
    als = AlternatingLeastSquares(factors, regularization , iterations)
    # Convert user-item matrix to sparse matrix
    sparse_user_item = csr_matrix(user_item.values)
    # Ensure that the matrix is in Compressed Sparse Row (CSR) format
    sparse_user_item = sparse_user_item.tocsr()
    # Fit the model
    als.fit(sparse_user_item)
    # Obtain user and item factors
    user_factors = als.user_factors
    item_factors = als.item_factors
    # Complete the original user-item matrix
    completed_user_item_matrix = user_factors.dot(item_factors.T)
    # Create a DataFrame from the completed matrix
    self.user_item_df = pd.DataFrame(completed_user_item_matrix, index=user_item.index, columns=user_item.columns)

    # return self.user_item_df


  def predict_als(self,user,item):
    return self.avg_rating + self.bi[item] + self.bu[user] + self.user_item_df.loc[user][item]

  def recommend(self, user_id, n,data):
    predicted_rating=[]
    m_id=data.ratings_data[data.ratings_data["userId"]==1]["movieId"].tolist()

    for item in data.user_item.columns:
        if item not in m_id:
          pred=self.predict_als(int(user_id),int(item))
          if pred > 5:
            pred=5
          elif pred < 0:
            pred = 0
          pred_dict={"item":item ,"score":pred}
          predicted_rating.append(pred_dict)
    predicted_rating=pd.DataFrame(predicted_rating)
    return predicted_rating.sort_values(by='score',ascending=False).head(n)
```

#Query Aspect Diversification

This class is use to find Diversity Score using QUAD & RxQUAD

```python
import math
class QUAD:
  def __init__(self,user,RL,data):
    self.RL=RL
    self.calculate_users_interest(user,data)

  # Method to find User Interest
  def calculate_users_interest(self,user,data):
    user_interec = data.ratings_data[data.ratings_data["userId"]==user]
    user_interec=user_interec[user_interec["rating"]>3]
    aspect_counts={}
    total_rating = len(user_interec)

    # Calculate total number of count of genre in user profile
    for _,row in user_interec.iterrows():
      movie=data.movies_data[data.movies_data["movieId"]==row["movieId"]]["genres"]
      movie_genres=movie.iloc[0].split('|')

      for genre in movie_genres:
        aspect_counts[genre]=aspect_counts.get(genre,0)+1

    # Calculate p(a|u)
    self.user_interest_distribution = {aspect: count / total_rating for aspect, count in aspect_counts.items()}

  # Indicator Function return True or False
  def indicator_func(self,genres,aspect):
    if aspect in genres:
      return True
    else:
      return False
  # Method To find P(i|u,a)
  def get_item_given_userAspect(self,item,genres,aspect):
    item_rel_aspect=0

    if self.indicator_func(genres,aspect):
        total_dino=0
        for item2, row in self.RL.iterrows():
          j_genre=self.RL.loc[item2]["genre"].split('|')
          if self.indicator_func(j_genre,aspect):
            total_dino+=row["score"]
        item_rel_aspect+=(self.RL.loc[item]["score"]/total_dino)
    return item_rel_aspect

  # Method to find Novelty Score using Quad Formulation
  def calc_item_rel_QUAD(self,item):
      Item_QUAD_rel=0
      # print("Item : ",item)
      i_genre=self.RL.loc[item]["genre"].split('|')
      for key in self.user_interest_distribution:
        aspect=key
        # Calculating Eq 1 p(S|u) * p(i|u,S)
        P_i_ua=self.get_item_given_userAspect(item,i_genre,aspect)
        print("item",item,"genres",i_genre,"aspect",aspect,"P_i_ua",P_i_ua)

        eq1= self.user_interest_distribution[key] * P_i_ua

        # Calculating * (1- P(j|u,s))
        eq2=1
        for itemj, row in self.RL.iterrows():
          j_genre=self.RL.loc[itemj]["genre"].split('|')
          P_j_ua=self.get_item_given_userAspect(itemj,j_genre,aspect)
          eq2*= (1- P_j_ua)
        # Calculating sumation eq1 * eq2
        Item_QUAD_rel+= eq1 * eq2
      return Item_QUAD_rel

  def calc_QUAD_rel_div(self):
    self.RL["QUAD_score"]=0
    for item,row in self.RL.iterrows():
      self.RL.loc[item,"QUAD_score"]=self.calc_item_rel_QUAD(item)


###### RxQUAD
  # Method to find Novelty Score using RxQuad Formulation
  def calc_item_rel_RxQUAD(self,item,alpha=0.5):
    stoping_probability=1-alpha
    Item_RxQUAD_rel=0
    # print("Item : ",item)
    i_genre=self.RL.loc[item]["genre"].split('|')
    for key in self.user_interest_distribution:
      aspect=key
      # Calculating Eq 1 p(S|u) * p(rel|i,u,S)
      if self.indicator_func(i_genre,aspect):
        P_rel_iua=self.RL.loc[item]["score"]/5
        eq1=self.user_interest_distribution[key]* P_rel_iua

        # Calculating * (1- P(rel|j,u,s))
        eq2=1
        for itemj, row in self.RL.iterrows():
          j_genre=self.RL.loc[itemj]["genre"].split('|')
          if self.indicator_func(j_genre,aspect):
            P_rel_jua=self.RL.loc[itemj]["score"]/5
            eq2*= (1- (P_rel_jua*stoping_probability))
        # Calculating sumation eq1 * eq2
        Item_RxQUAD_rel+= eq1 * eq2
    return Item_RxQUAD_rel

  def calc_RxQUAD_rel_div(self):
    self.RL["RxQUAD_score"]=0
    for item,row in self.RL.iterrows():
      self.RL.loc[item,"RxQUAD_score"]=self.calc_item_rel_RxQUAD(item)

```

#IB+ Recommender (Used to only create subrofile)

This class is use to genrate RS  using IB Recommender to find Subprofiles for user

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
class IB_Recommender:
  def __init__(self):
    self.items_similarities=None

  def calc_similarity(self,user_item):
    self.items_similarities=cosine_similarity(user_item.T)
    np.fill_diagonal(self.items_similarities, 0)
    self.items_similarities=pd.DataFrame(self.items_similarities,index=user_item.columns,columns=user_item.columns)

  def recommend(self,user, user_profile,data, n=20):
    up=data.ratings_data[data.ratings_data["userId"]==user]

    scores = pd.DataFrame(np.dot(user_profile, self.items_similarities),index= self.items_similarities.columns,columns=["score"])
    scores=scores[~scores.index.isin(up["movieId"])]
    return scores.sort_values(by="score",ascending=False).head(n)

  def get_sim(self,item,n=100):
    return pd.DataFrame(pd.Series(self.items_similarities[item],name="score")).head(20)


```

#SPAD

This class is use to find Diversity Score using SPAD & RxSPAD

```python
import math

class SPAD:
  def __init__(self,RL_ALS,subprofile_size=10):
    self.subprofile_size=subprofile_size
    self.subprofiles_total_element=0
    self.RL=None
    self.RL_ALS=RL_ALS
  # Method to Extract & refine Subprofile
  def extract_subprofiles(self,user_id,user_profile,RL,items_similarities):
    # Finding similarities Score between users profile and Recommendation List
    self.RL=RL
    self.get_up_similar_items(user_profile,items_similarities)
    # Genrate neighbors
    self.genrate_neighbors()
    # Refine _subprofile
    self.refine_subprofile()

  #Method to find items most similar to user past interations
  def get_up_similar_items(self,user_profile,items_similarities):
    # Finding similarities Score between users profile and Recommendation List
    self.up_similar_items=items_similarities[user_profile.index]
    self.up_similar_items=self.up_similar_items[self.up_similar_items.index.isin(self.RL.index)]

  # method to genrate neighbours for all the user 's past interaction
  def genrate_neighbors(self):
    # Genrate neighbors
    self.neighbors={}
    for item in self.up_similar_items.columns:
      self.neighbors[item]=[]
      for i in self.up_similar_items[[item]].sort_values(by=item,ascending=False).head(self.subprofile_size).index:
        self.neighbors[item].append(i)

  # Method to refine all the subprofile
  def refine_subprofile(self):
    # Generate candidate subprofiles based on neighbors
    candidate_subprofiles = []
    for movie_id, neighbors in self.neighbors.items():
        subprofile =  list( [movie_id]+neighbors)
        candidate_subprofiles.append(subprofile)

    self.refined_subprofiles=[]

    for subprofile in candidate_subprofiles:
      if len(self.refined_subprofiles)==0:
        self.refined_subprofiles.append(subprofile)
      else:
        subprofile2=subprofile
        for item1 in subprofile:
          item_to_remove=[]
          for item2 in subprofile2:
            if item1!=item2:
              if any(set({item1,item2}).issubset(existing_subprofile) for existing_subprofile in self.refined_subprofiles):
                item_to_remove.append(item2)
          for i_remove in list(set(item_to_remove)):
            subprofile2.remove(i_remove)
        self.refined_subprofiles.append(subprofile2)

  # Method to Calculte total Element present in all the subprofile
  def calc_subprofiles_total_element(self):
    for subprofile_temp in self.refined_subprofiles:
      self.subprofiles_total_element+=len(subprofile_temp)
    # print("self.subprofiles_total_element",self.subprofiles_total_element)

  # Method to find user's Interest for every Subprofile
  def calculate_users_interest(self):
    self.calc_subprofiles_total_element()
    self.subprofile_userinterest={}
    for i in range(len(self.refined_subprofiles)):
      subprofile=self.refined_subprofiles[i]
      self.subprofile_userinterest[i]={"subprofile":subprofile,"score":len(subprofile)/self.subprofiles_total_element}

  # Indicator function returns True if item present in the subprofile else return false
  def indicator_func(self,item,subprofile):
    if item in subprofile["subprofile"]:
      return True
    else:
      return False
  # Method to calc P(i|u,s)
  def get_item_given_userAspect(self,item,subprofile):
    item_rel_aspect=0
    if self.indicator_func(item,subprofile):
        total_dino=0
        for item2, row in self.RL_ALS.iterrows():
          if self.indicator_func(item2,subprofile):
            total_dino+=row["score"]
        item_rel_aspect+=(self.RL_ALS.loc[item]["score"]/total_dino)
    return item_rel_aspect
  # Mhtod to find SPAD(i,RL)
  def calc_item_rel(self,item):
    Item_SPAD_rel=0
    # print("Item : ",item)
    for key in range(len(self.refined_subprofiles)):
      subprofile=self.subprofile_userinterest[key]
      # print("subprofile : ",subprofile,end=" ")
      # Calculating Eq 1 p(S|u) * p(i|u,S)
      P_i_ua=self.get_item_given_userAspect(item,subprofile)
      # print("item",item,"subprofile",subprofile,"P_i_ua",P_i_ua)
      eq1= subprofile["score"] * P_i_ua

      # Calculating * (1- P(j|u,s))
      eq2=1
      for itemj, row in self.RL_ALS.iterrows():
        P_j_ua=self.get_item_given_userAspect(itemj,subprofile)
        eq2*= (1- P_j_ua)
      # print("P_i_ua : ",P_i_ua,"eq1 : ",eq1 ,"eq2:",eq2)
      # Calculating sumation eq1 * eq2
      Item_SPAD_rel+= eq1 * eq2
    return Item_SPAD_rel


  def calc_SPAD_rel_div(self):
    self.RL_ALS["SPAD_SCORE"]=0
    for item,row in self.RL_ALS.iterrows():
      self.RL_ALS.loc[item,"SPAD_SCORE"]=self.calc_item_rel(item)
      # print(self.RL.loc[item,"rel_score"])

##### RSPAD
  # Method to Find RxSPAD(i,RL)
  def calc_item_rel_RSPAD(self,item,alpha=0.5):
    stoping_probability=1-alpha
    Item_RSPAD_rel=0
    # print("Item : ",item)
    for key in range(len(self.refined_subprofiles)):
      subprofile=self.subprofile_userinterest[key]
      # print("subprofile : ",subprofile,end=" ")
      # Calculating Eq 1 p(S|u) * p(rel|i,u,S)
      P_rel_iua=self.RL_ALS.loc[item]["score"]/5
      # print(P_rel_iua)
      eq1= subprofile["score"] * P_rel_iua

      # Calculating * (1- P(rel|j,u,s))
      eq2=1
      for itemj, row in self.RL_ALS.iterrows():
        P_rel_jua=self.RL_ALS.loc[itemj]["score"]/5
        eq2*= (1- (P_rel_jua*stoping_probability))
      # print("P_i_ua : ",P_i_ua,"eq1 : ",eq1 ,"eq2:",eq2)
      # Calculating sumation eq1 * eq2
      Item_RSPAD_rel+= eq1 * eq2
    return Item_RSPAD_rel



  def calc_RSPAD_rel_div(self):
    self.RL_ALS["RSPAD_SCORE"]=0
    for item,row in self.RL_ALS.iterrows():
      self.RL_ALS.loc[item,"RSPAD_SCORE"]=self.calc_item_rel_RSPAD(item)
      # print(self.RL_ALS.loc[item,"rel_score"])

```
